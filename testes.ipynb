```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the CSV file into a DataFrame
df = pd.read_csv("pizzas.csv")
df
```

```python
# Create a scatter plot to visualize the relationship between diameter and price
df.plot(kind="scatter", x="diameter", y="price")

plt.xlabel("Diameter (cm)")
plt.ylabel("Price (BRL)")
plt.title("Relationship between Pizza Diameter and Price")

plt.show()
# Based on the scatter plot, we can see that there is a positive correlation between the diameter of the pizza and its price. As the diameter increases, the price also tends to increase. 
# This suggests that larger pizzas are generally more expensive than smaller ones.
```

```python
from sklearn import linear_model

# Create a linear regression model and fit it to the data
model = linear_model.LinearRegression()
x = df[["diameter"]]
y = df["price"]
# Fit the model to the data
model.fit(x, y)
```

```python
# Now we can use the model to make predictions. For example, if we want to predict the price of a pizza with a diameter of 30 cm, we can do it like this:
print(model.predict([[110]])[0])
```
